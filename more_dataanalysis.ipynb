## Exploration of data set


```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb

PATH = 'C:\\Users\\simon\\PycharmProjects\\covid\\raw_data\\'
```

```python
df = pd.read_csv(PATH+'covid_de.csv', parse_dates=['date'])  # dataset from RKI (cleaned on kaggle)
```

```python
df.loc[df['gender'].isnull(), 'gender'] = 'unknown'
df.loc[df['age_group'].isnull(), 'age_group'] = 'unknown'
```

```python
df.sort_values(by='date', inplace=True)
```

```python
x_dates = df['date'].dt.strftime('%d/%m').unique()
```

```python
fig = plt.figure()
sb.barplot(data=df, x='age_group', y='cases', hue='gender', estimator=sum, ci=None, order=['00-04', '05-14', '15-34', '35-59', '60-79', '80-99', 'unknown'])
plt.gca().set_yscale('log')
```
